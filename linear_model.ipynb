```python
x_train=df[['YearsExperience']]
y_train=df[['Salary']]
```

```python
x_train=df[['YearsExperience']]
y_train=df[['Salary']]
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df["pred_salary"]=model.predict(x_train)
```

```python
df.head()
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,df['pred_salary'],color='red')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(df['YearsExperience'],df['pred_salary'])
mse
```

```python
from sklearn.metrics import r2_score
r2=r2_score(df['Salary'],df['pred_salary'])
r2
```
